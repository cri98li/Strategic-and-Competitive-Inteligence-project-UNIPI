```python
import pandas as pd
import tqdm.auto as tqdm
import numpy as np
```

```python
import json

with open("../data/ipc_class_dict.json") as json_file:
    ipc_class_dict = json.load(json_file)
```

```python
df = pd.read_csv("../data/subest1_filtrato.csv")
```

```python
df.title = df.title.str.lower()
df.abstract = df.abstract.str.lower()
df.claims = df.claims.str.lower()
```

```python
keywords_cosmetics = ["cosmetic",
                      "beauty", 
                      "body", 
                      "skin", 
                      "nail", 
                      "fingernail", 
                      "cleansing", 
                      "manicur", #manicure and manicuring
                      "varnish", 
                      "polish",
                      "nail polish"
                     ]

keywords_other = ["molecule",
                  "particle", 
                  "pharmacological",
                  "technolog",
                  "nanotechnolog",
                  "altering"
                 ]
```

```python
for key in keywords_cosmetics:
    df[key] = df.title.str.count(key).fillna(0) + df.abstract.str.count(key).fillna(0) + df.claims.str.count(key).fillna(0)
    
for key in keywords_other:
    df[key] = df.title.str.count(key).fillna(0) + df.abstract.str.count(key).fillna(0) + df.claims.str.count(key).fillna(0)
```

```python
df["total_cosmetics"] = 0
for key in keywords_cosmetics:
    df["total_cosmetics"] += df[key]
    
df["total_other"] = 0
for key in keywords_other:
    df["total_other"] += df[key]
```

```python
location_interesting_patents = [
    2571,
    3341,
    20194,
    23331
]
```

```python
ipc_count = {}

for loc in location_interesting_patents:
    for ipc_class in df.loc[loc]["ipc_classes"].split(","):
        ipc_class = ipc_class[:ipc_class.index("/")]+"/00"
        
        if ipc_class in ipc_count:
            ipc_count[ipc_class] +=1
        else:
            ipc_count[ipc_class] = 1

for k, v in sorted(ipc_count.items(), key=lambda x: -x[1]):
    if k in ipc_class_dict:
        print(f"{v}, {ipc_class_dict[k][:100]} ({k})\n")
    else:
        print(f"{v}, {k}\n")
```
